```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math as m
```

```python
data=pd.read_csv("boston_train.csv")
x=data.drop("medv",1).values
y=data["medv"].values
length=len(data.columns)-1
```

```python
def pred(w):
    y_pred=[]
    for i in range(len(y)):
        temp=w[0]
        for j in range(1,len(w)):
            temp=temp+x[i,j-1]*w[j]
        y_pred.append(temp)
    return y_pred
```

```python
def cost(y_pred):
    m=len(x)
    cost=0
    for i in range(len(x)):
        cost=cost+((y_pred[i]-y[i])**2) 
    cost=(1/(2*m))*cost
    return cost
```

```python
WT=[]
for i in range(length):
    WT.append(np.ndarray.tolist(np.random.randint(1000,size=14)*0.0001))
```

```python
Y_PredT=[]
for i in range(length):
    Y_PredT.append(pred(WT[i]))
```

```python
Tcost=[]
for i in range(length):
    Tcost.append(cost(Y_PredT[i]))
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
for i in range(length):
    ax.scatter(WT[i], Tcost)
ax.set_xlabel('wights')
ax.set_ylabel('cost')
ax.set_title('scatter plot')
plt.show()
```
